
# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
# Write your code here

import pandas as pd

data = {
    "Name": ["Aman", "Rahul", "Neha", "Priya", "Ravi"],
    "Marks": [21, 22, 20, 23, 24],
    "City": ["bpl", "indr", "bpl", "bpl", "bpl"]
}

df = pd.DataFrame(data)
df["City"] = df["City"].replace("bpl", "Bhopal")
df["City"] = df["City"].replace("indr", "Indore")

print(df)

````


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
# Write your code here
import pandas as pd

data = {
    "Name": ["Aman", "Rahul", "Aman", "Neha"]
}

df = pd.DataFrame(data)
print("Duplicate Rows:")
print(df.duplicated())
df2 = df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(df2)
````
