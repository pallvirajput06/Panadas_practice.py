# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
# Write your code here
import numpy as np
import pandas as pd
df = pd.DataFrame({
    "City" : ["Bhopal", "Indore", "Bhopal", "Delhi"]
})
df["City"].unique()
````


# Q7. Filter Records

## Problem Statement
Display students having marks greater than 75.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
import pandas as pd

df = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Simran"],
    "Marks": [68, 72, 75, 80, 70, 74]
})
filtered_df = df[df["Marks"] > 75]
print(filtered_df)
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
import pandas as pd

df = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Simran"],
    "Marks": [68, 72, 75, 80, 70, 74],
    "Age": [17, 19, 21, 23, 25, 18]
})

filtered_df = df[(df["Marks"] > 70) & (df["Age"] > 20)]

print(filtered_df)
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
# Write your code here
import pandas as pd

df = pd.DataFrame({
    "Employee": ["Amit", "Neha", "Rahul"],
    "Department": ["IT", "HR", "IT"],
    "Salary": [60000, 50000, 60000]
})

total_salary_by_department = df.groupby("Department")["Salary"].sum()
print(total_salary_by_department)
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
# Write your code here
import pandas as pd

df = pd.DataFrame({
    "Name": ["Amit", "Neha", "Rahul", "Priya", "Karan", "Simran"],
    "Age": [18, 20, 19, 21, 22, 23],
    "City": ["Bhopal", "Indore", "Delhi", "Bhopal", "Indore", "Delhi"],
    "Marks": [85, 78, 92, 88, 75, 81]
})

print("Original DataFrame:")
print(df)
print(df["City"].unique())
print(df["City"].value_counts())
print(df[df["Marks"] > 80])
print(df.sort_values("Marks"))
````
